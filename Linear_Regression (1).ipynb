```python
x_train=df_train.drop(['target'],axis=1).values
y_train=df_train['target'].values
print("x_train=",x_train)
print("y_train=",y_train)
```

```python
df_test=pd.read_csv('/content/Lineardata_test.csv')
df_test
```

```python
X_train_1=mean_normalized_data_train.drop(['target'],axis=1).values
Y_train_1=mean_normalized_data_train['target'].values
```

```python
X_test=mean_normalized_data_test.drop(['ID'],axis=1).values
X_test
```

```python
# Split the data into training and validation sets
train_ratio = 0.8
spilt_index = int(train_ratio * len(X_train_1))
```

```python
X_train = X_train_1[:spilt_index]
X_val=X_train_1[spilt_index:]   # X_val represents the values for valdiation
Y_train = Y_train_1[:spilt_index]
Y_val=Y_train_1[spilt_index:]
X_val.shape
```

```python
ones_train=np.ones([len(X_train),1])
x_train = np.concatenate((ones_train, X_train),axis=1)
ones_val=np.ones([len(X_val),1])
x_val = np.concatenate((ones_val, X_val),axis=1)
ones_test=np.ones([len(X_test),1])
x_test = np.concatenate((ones_test, X_test),axis=1)
```

```python
# Calculate cost function
def compute_cost(X_train, y_train, theta):
    cost=0
    m = len(Y_train)
    predictions = X_train.dot(theta)
    cost = cost+(1/(2*m)) * np.sum(np.square(predictions - Y_train))
    return cost
```

```python
# Calculate gradient
def compute_gradient(X_train, Y_train, theta):
    m = len(Y_train)
    predictions = X_train.dot(theta)
    errors = predictions - Y_train
    gradient = (1/m) * X_train.T.dot(errors)
    return gradient
```

```python
def gradient_descent(X_train, Y_train, learning_rate=0.3, iters=10000):
    theta = np.zeros(X_train.shape[1])
    cost=[]
    for i in range(iters):
        gradient=compute_gradient(X_train, Y_train, theta)
        theta=theta-learning_rate*gradient
        c=compute_cost(X_train,y_train,theta)
        cost.append(c)
        if i%1000==0:
            print("Iteration=",i)
            print("Cost=", c)
    return theta,cost
```

```python
theta,cost=gradient_descent(x_train, Y_train, learning_rate=0.3, iters=10000)
print(theta)
```

```python
plt.plot(np.arange(len(cost)),cost)
plt.xlabel('Number of Iterations')
plt.ylabel('Cost')
```

```python
def predict(x,theta):
    pred=x.dot(theta)
    return pred
```

```python
Y_train_pred=predict(x_train,theta)
print("Y_train_pred=", Y_train_pred)
print("Y_train=", Y_train)
```

```python
plt.scatter(x_train[:,1],Y_train,color="orange",label='True')
plt.scatter(x_train[:,1],Y_train_pred,marker="*",label='Predicted')
plt.xlabel("Few training examples")
plt.ylabel("Label")
plt.legend()
plt.show
```

```python
Y_val_pred=predict(x_val,theta)
print("Y_val_pred=", Y_val_pred)
print("Y_val=", Y_val)
```

```python
Y_test_pred=x_test.dot(theta)
print("Y_test_pred=", Y_test_pred)
```

```python
# R-squared score
ssr = np.sum((Y_val- Y_val_pred)**2)  # sum of squared residuals(ssr)
sst = np.sum((Y_val - np.mean(Y_val))**2)  # Total Sum of Squares(sst)
r_squared = 1 - (ssr / sst)
print("R-squared score of validation set=",r_squared)
```
